```python
import gensim
import os
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

stop_words = stopwords.words('english')

story = []
for filename in os.listdir('/content/data'):

    f = open(os.path.join('/content/data',filename), encoding='latin-1')
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        tokenized_sent = simple_preprocess(sent)
        filtered_sent = [word for word in tokenized_sent if word not in stop_words]  # Removing stop words.
        story.append(filtered_sent)
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
